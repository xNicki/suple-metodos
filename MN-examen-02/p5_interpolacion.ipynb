```python
# Derivadas parciales para regresión lineal
# ####################################################################
def der_parcial_4(xs: list, ys: list) -> tuple[float, float, float]:
   
    c_4 = 0
    c_3 = 0
    c_2 = 0
    c_1 = 0
    c_0 = 0
    c_ind = 0
    for xi, yi in zip(xs, ys):
        # coeficiente del término independiente
        
        c_ind += xi**4 * yi
        
        c_4 += xi**8
        c_3 += xi**7
        c_2 += xi**6

        # coeficiente del parámetro 1
        c_1 += xi**5

        # coeficiente del parámetro 0
        c_0 += xi**4

    return (c_4, c_3, c_2, c_1, c_0, c_ind)

def der_parcial_3(xs: list, ys: list) -> tuple[float, float, float]:
   
    c_4 = 0
    c_3 = 0
    c_2 = 0
    c_1 = 0
    c_0 = 0
    c_ind = 0
    for xi, yi in zip(xs, ys):
        # coeficiente del término independiente
        
        c_ind += xi**3 * yi
        
        c_4 += xi**7
        c_3 += xi**6
        c_2 += xi**5

        # coeficiente del parámetro 1
        c_1 += xi**4

        # coeficiente del parámetro 0
        c_0 += xi**3

    return (c_4, c_3, c_2, c_1, c_0, c_ind)

def der_parcial_2(xs: list, ys: list) -> tuple[float, float, float]:
   
    c_4 = 0
    c_3 = 0
    c_2 = 0
    c_1 = 0
    c_0 = 0
    c_ind = 0
    for xi, yi in zip(xs, ys):
        # coeficiente del término independiente
        
        c_ind += xi**2 * yi
        
        c_4 += xi**6
        c_3 += xi**5
        c_2 += xi**4

        # coeficiente del parámetro 1
        c_1 += xi**3

        # coeficiente del parámetro 0
        c_0 += xi**2

    return (c_4, c_3, c_2, c_1, c_0, c_ind)

def der_parcial_1(xs: list, ys: list) -> tuple[float, float, float]:
   
    c_4 = 0
    c_3 = 0
    c_2 = 0
    c_1 = 0
    c_0 = 0
    c_ind = 0
    for xi, yi in zip(xs, ys):
        # coeficiente del término independiente
        
        c_ind += xi**1 * yi
        
        c_4 += xi**5
        c_3 += xi**4
        c_2 += xi**3

        # coeficiente del parámetro 1
        c_1 += xi**2

        # coeficiente del parámetro 0
        c_0 += xi**1

    return (c_4, c_3, c_2, c_1, c_0, c_ind)


def der_parcial_0(xs: list, ys: list) -> tuple[float, float, float]:
   
    c_4 = 0
    c_3 = 0
    c_2 = 0
    c_1 = 0
    c_0 = 0
    c_ind = 0
    for xi, yi in zip(xs, ys):
        # coeficiente del término independiente
        
        c_ind += xi**0 * yi
        
        c_4 += xi**4
        c_3 += xi**3
        c_2 += xi**2

        # coeficiente del parámetro 1
        c_1 += xi**1

        # coeficiente del parámetro 0
        c_0 += xi**0

    return (c_4, c_3, c_2, c_1, c_0, c_ind)
```
